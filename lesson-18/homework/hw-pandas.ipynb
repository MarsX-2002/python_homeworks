#### Part 3: Challenges and Explorations  
 
- From **`iris.json`**: Calculate the mean, median, and standard deviation for each numerical column.  
- From **`titanic.xlsx`**: Find the minimum, maximum, and sum of passenger ages.  

- From **`movie.csv`**:  
    - Identify the director with the highest total `director_facebook_likes`.  
    - Find the 5 longest movies and their respective directors.  

- From **Flights parquet file**:  
    - Check for missing values in the dataset. Fill missing values in a numerical column with the column’s mean.  

```python
# 1. iris.json
# Select only numerical columns
numerical_columns = df_iris.select_dtypes(include=['number']).columns

numerical_stats = df_iris[numerical_columns].describe().loc[['mean', 'std']] 
median_values = df_iris[numerical_columns].median()

# add median column to numerical stats
numerical_stats.loc['median'] = median_values

print(numerical_stats)
```

```python
# 2. titanic.xlsx
# passenger ages
age_min = df_titanic['Age'].min() 
age_max = df_titanic['Age'].max() 
age_sum = df_titanic['Age'].sum() 

# Display results
print(f"Minimum age: {age_min}")
print(f"Maximum age: {age_max}")
print(f"Sum of ages: {age_sum}")
```

```python
# 3. `movie.csv
# group by director_name
director_likes = df_movie.groupby('director_name')['director_facebook_likes'].sum()

# highest total likes
top_director  = director_likes.idxmax()
top_likes = director_likes.max()

print(f"The director with the highest total likes is {top_director} with {top_likes} likes.")
```

```python
# 4. Flights parquet file
# a) missing values 
missing_values_flights = flights.isnull().sum()
print(f"missing values: {missing_values_flights}")
```

```python
# b) fill missing values with mean
numerical_columns_flights = flights.select_dtypes(include=['number']).columns

flights[numerical_columns_flights] = flights[numerical_columns_flights].apply(lambda x: x.fillna(x.mean()))

print(flights.isnull().sum())
```
