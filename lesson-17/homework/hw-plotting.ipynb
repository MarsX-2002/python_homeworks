#### **1. Basic Plotting**
- **Task**: Plot the function $ f(x) = x^2 - 4x + 4 $ for $ x $ values between -10 and 10. Customize the plot with appropriate labels for the axes and a title.

```python
x = np.linspace(-10, 10, 1000)  # x values between -10 and 10, with 1000 points
y = x ** 2 - 4 * x + 4  # y = f(x)

plt.plot(x, y, label="$ f(x) = x^2 - 4x + 4 $")
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.show()
```

#### **2. Sine and Cosine Plot**
- **Task**: Plot $ \sin(x) $ and $ \cos(x) $ on the same graph for $ x $ values ranging from 0 to $ 2\pi $. Use different line styles, markers, and colors to distinguish between the two functions. Add a legend.


```python
sin_func = np.sin
cos_func = np.cos

x = np.linspace(0, 2 * np.pi, 100)  # x values between 0 and 2pi, with 100 points

y_sin = sin_func(x)
y_cos = cos_func(x)

plt.plot(x, y_sin, 'r-o', label="sin(x)")  # 'r-' red solid line and 'o' markers
plt.plot(x, y_cos, 'b--s', label="cos(x)")  # 'b--' blue dashed line and 's' markers


plt.xlabel('x')
plt.ylabel('y(x)')

# Customize x-ticks to display multiples of 
ticks = [0, np.pi / 2, np.pi, 3 * np.pi / 2, 2 * np.pi]
tick_labels = [
    r"$0$", r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$"
]
plt.xticks(ticks, tick_labels)

plt.legend()
plt.grid(True)
plt.show()
```

#### **3. Subplots**
- **Task**: Create a 2x2 grid of subplots. In each subplot, plot:
  - Top-left: $ f(x) = x^3 $
  - Top-right: $ f(x) = \sin(x) $
  - Bottom-left: $ f(x) = e^x $
  - Bottom-right: $ f(x) = \log(x+1) $ (for $ x \geq 0 $)

  Customize each plot with titles, axis labels, and different colors.

```python
# top left
x1 = np.linspace(-10, 10, 100)
y1 = x1 ** 3

# top right
x2 = np.linspace(0, 2 * np.pi, 100)  # x values between 0 and 2pi, with 100 points
y2 = np.sin(x2)

# bottom left
x3 = np.linspace(-2, 2, 100)
y3 = np.exp(x3)

# bottom right
x4 = np.linspace(0, 10, 100)
y4 = np.log10(x4 + 1)

# top left
plt.subplot(2, 2, 1)
plt.plot(x1, y1, 'r-')
plt.title('$ f(x) = x^3 $')
plt.xlabel('x')
plt.ylabel('y(x)')

# top right
plt.subplot(2, 2, 2)
plt.plot(x2, y2, 'b-')
plt.title(r'$ f(x) = \sin(x) $')
plt.xlabel('x')
plt.ylabel('y(x)')
# Customize x-ticks to display multiples of 
ticks = [0, np.pi / 2, np.pi, 3 * np.pi / 2, 2 * np.pi]
tick_labels = [
    r"$0$", r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$"
]
plt.xticks(ticks, tick_labels)

# bottom left
plt.subplot(2, 2, 3)
plt.plot(x3, y3, 'g-')
plt.title('$ f(x) = e^x $')
plt.xlabel('x')
plt.ylabel('y(x)')

# bottom right
plt.subplot(2, 2, 4)
plt.plot(x4, y4, 'y-')
plt.title(r'$ f(x) = \log_{10}(x+1) \ (x \geq 0) $') 
plt.xlabel('x')
plt.ylabel('y(x)')

# show
plt.tight_layout()
plt.show()
```

#### **4. Scatter Plot**
- **Task**: Create a scatter plot of 100 random points in a 2D space. The x and y values should be randomly generated from a uniform distribution between 0 and 10. Use different colors and markers for the points. Add a title, axis labels, and a grid.

```python
# Generate random data
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

# Create scatter plot
plt.scatter(x, y, c=np.random.rand(100), marker='o', cmap='viridis', edgecolor='k')
plt.title('Scatter Plot of Random Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.colorbar(label='Color Intensity')
plt.show()
```

#### **5. Histogram**
- **Task**: Generate a random dataset of 1000 values sampled from a normal distribution (mean=0, std=1). Plot a histogram of the data with 30 bins. Add a title and axis labels. Adjust the transparency of the bars using the `alpha` parameter.

```python
# Generate random data
data = np.random.normal(0, 1, 1000)

# Plot histogram
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

#### **6. 3D Plotting**
- **Task**: Create a 3D surface plot for the function $ f(x, y) = \cos(x^2 + y^2) $ over the range of $ x $ and $ y $ values from -5 to 5. Use a suitable colormap and add a colorbar. Set appropriate labels for the axes and title.

```python
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.cos(x**2 + y**2)

# Create 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Set labels
ax.set_title(r'3D Surface Plot: $f(x, y) = \cos(x^2 + y^2)$')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

#### **7. Bar Chart**
- **Task**: Create a vertical bar chart displaying the sales data for five different products: `['Product A', 'Product B', 'Product C', 'Product D', 'Product E']`. The sales values for each product are `[200, 150, 250, 175, 225]`. Customize the chart with a title, axis labels, and different bar colors.

```python
# Sales data
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]
colors = ['red', 'blue', 'green', 'purple', 'orange']

# Create bar chart
plt.bar(products, sales, color=colors, edgecolor='black')
plt.title('Sales Data by Product')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

#### **8. Stacked Bar Chart**
- **Task**: Create a stacked bar chart that shows the contribution of three different categories (`'Category A'`, `'Category B'`, and `'Category C'`) over four time periods (`'T1'`, `'T2'`, `'T3'`, `'T4'`). Use sample data for each category at each time period. Customize the chart with a title, axis labels, and a legend.

```python
# Data
time_periods = ['T1', 'T2', 'T3', 'T4']
category_a = [5, 7, 8, 6]
category_b = [4, 6, 7, 5]
category_c = [6, 5, 8, 7]

# Create stacked bar chart
bar_width = 0.6
plt.bar(time_periods, category_a, color='red', edgecolor='black', label='Category A')
plt.bar(time_periods, category_b, bottom=category_a, color='blue', edgecolor='black', label='Category B')
plt.bar(time_periods, category_c, bottom=np.array(category_a) + np.array(category_b), color='green', edgecolor='black', label='Category C')

plt.title('Stacked Bar Chart')
plt.xlabel('Time Periods')
plt.ylabel('Values')
plt.legend()
plt.show()
```
